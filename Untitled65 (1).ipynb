## **##Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**
##Answer:
Anomaly detection is the process of identifying rare items, events, or observations in a dataset that deviate significantly from the norm or expected patterns. These anomalies (also called outliers) can indicate fraud, defects, or novel discoveries. It's widely used in fields like cybersecurity, finance, and healthcare. Anomalies are detected using statistical methods, machine learning algorithms, or domain-specific rules.

There are three main types of anomalies:

- Point Anomalies: These are individual data points that stand out from the rest of the dataset, regardless of context. They are the simplest type.
Example: In a dataset of daily credit card transactions where most are $10–$100, a single transaction of $10,000 could be a point anomaly, potentially indicating fraud.

- Contextual Anomalies: These depend on the context (e.g., time, location) and are anomalous only within a specific situation, even if they seem normal otherwise.
Example: A temperature reading of 30°C might be normal in summer but anomalous in winter (context: season). In time series data, a sudden spike in website traffic at 3 AM (when it's usually low) could be contextual.

- Collective Anomalies: These occur when a group of data points collectively deviate from the norm, even if individual points seem normal.
Example: In network traffic data, a sequence of small, frequent packets might not be anomalous individually, but together they could indicate a DDoS attack. Another example is a series of ECG readings that form an irregular pattern, signaling a heart anomaly.

Understanding these types helps choose appropriate detection methods, like statistical tests for point anomalies or clustering for collective ones.

## **##Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**
# Answer:
Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are popular unsupervised anomaly detection algorithms. They differ in how they identify outliers, scalability, and applicability. Below is a comparison:

- Isolation Forest:

- Approach: This ensemble method (based on random forests) isolates anomalies by randomly partitioning the data. Anomalies are easier to isolate (require fewer partitions) than normal points, as they are farther from dense regions. It uses an anomaly score based on path length in isolation trees.

- Suitable Use Cases: High-dimensional datasets with a small percentage of anomalies, like fraud detection in large transaction logs or intrusion detection in networks. It's efficient for big data (scales well with size) but assumes anomalies are few and different.

- Pros/Cons: Fast and low memory usage; not great for datasets with many anomalies or local outliers.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise):

- Approach: A clustering algorithm that groups dense regions and labels low-density points as noise (anomalies). It uses parameters like epsilon (neighborhood radius) and minPts (minimum points for a core point) to define clusters; points not in any cluster are outliers.

- Suitable Use Cases: Spatial or geospatial data, such as detecting unusual locations in GPS tracks or anomalies in image segmentation. Ideal for datasets with varying densities and arbitrary-shaped clusters.

Pros/Cons: Handles noise well and doesn't require specifying cluster count; sensitive to parameter tuning and less effective in high dimensions.

Local Outlier Factor (LOF):

- Approach: Measures the local density deviation of a point compared to its neighbors. It computes a score where points with much lower density than their k-nearest neighbors are outliers, capturing local anomalies.

- Suitable Use Cases: Datasets where anomalies are relative to local contexts, like detecting fraud in credit scores (where outliers vary by region) or anomalies in sensor data from IoT devices. Good for mixed-density data.

- Pros/Cons: Effective for local outliers and doesn't assume global distribution; computationally intensive for large datasets.

In summary, choose Isolation Forest for scalability, DBSCAN for clustering with noise, and LOF for local density-based detection. For Python implementation, all are available in Scikit-learn.

## **##Question 3: What are the key components of a Time Series? Explain each with one example.**

# Answer:
A time series is a sequence of data points collected at successive time intervals. It has four key components: trend, seasonality, cyclicity, and residuals (or irregularity/noise). Decomposing a time series helps analyze and forecast it.

- Trend: The long-term direction or pattern in the data (upward, downward, or stable).
Example: In stock market data, a consistent increase in a company's share price over years due to business growth represents an upward trend.

- Seasonality: Regular, repeating patterns at fixed intervals (e.g., daily, monthly).
Example: Higher ice cream sales every summer (repeating annually) in retail data.

- Cyclicity: Fluctuations that occur over longer, irregular periods, often tied to economic cycles.
Example: Boom-and-bust cycles in housing prices every 7–10 years due to economic factors.

- Residuals (Irregularity/Noise): Random variations not explained by the other components, often due to unpredictable events.
Example: A sudden drop in airline passengers due to a one-time event like a natural disaster.

These components can be additive (e.g., value = trend + seasonality + residuals) or multiplicative, and tools like Python's statsmodels can decompose them.

## **##Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**

# Answer:
Stationarity in a time series means the statistical properties (mean, variance, autocorrelation) remain constant over time. A stationary series has no trend or seasonality, making it easier to model and forecast (e.g., for ARIMA models).

. Testing for Stationarity:

- Visual Tests: Plot the series and check for trends or seasonality. Use rolling statistics (e.g., moving average) to see if mean/variance changes.

- Statistical Tests: Augmented Dickey-Fuller (ADF) test (null hypothesis: non-stationary; p-value < 0.05 indicates stationarity) or KPSS test.
Example in Python: Use statsmodels.tsa.stattools.adfuller(series).

. Transforming Non-Stationary to Stationary:

- Differencing: Subtract previous values to remove trends (e.g., first-order: series.diff()).

- Log Transformation: Stabilize variance (e.g., np.log(series)).

Detrending/Deseasonalizing: Subtract trend or divide by seasonal component.
Example: For a trending series like stock prices, apply differencing until ADF confirms stationarity.

These steps ensure models like ARIMA perform accurately.

# **##Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.**

# Answer:
These are autoregressive models for time series forecasting, building on each other:

- AR (Autoregressive): Models future values based on past values (lags). Structure: y_t = c + φ1y_{t-1} + ... + φpy_{t-p} + ε_t (p = order). Application: Series with autocorrelation, like predicting temperature based on previous days.

- MA (Moving Average): Models based on past errors (white noise). Structure: y_t = c + θ1ε_{t-1} + ... + θqε_{t-q} + ε_t (q = order). Application: Smoothing random shocks, e.g., in financial noise.

- ARIMA (Autoregressive Integrated Moving Average): Combines AR and MA with differencing for non-stationary data. Structure: ARIMA(p,d,q) where d = differencing order. Application: General forecasting, like stock prices.

- SARIMA (Seasonal ARIMA): Extends ARIMA with seasonal components. Structure: SARIMA(p,d,q)(P,D,Q)m where m = seasonal period. Application: Seasonal data, e.g., monthly sales.

- SARIMAX: Adds exogenous variables to SARIMA. Structure: Includes external factors like weather. Application: Forecasting with influences, e.g., energy demand affected by temperature.

Choose based on data patterns: AR/MA for simple, ARIMA for trends, SARIMA/X for seasonality/external factors.

```python
##Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components (Include your Python code and output in the code box below.)

##python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.datasets as datasets

# Load AirPassengers dataset
data = pd.read_csv('AirPassengers (2).csv')
df = pd.DataFrame(data)
df['Month'] = pd.to_datetime(df['Month']) # Convert 'Month' column to datetime
df.set_index('Month', inplace=True)
series = df['#Passengers']  # Use the correct column name

# Plot original series
plt.figure(figsize=(10, 4))
plt.plot(series, label='Original')
plt.title('Original AirPassengers Series')
plt.legend()
plt.show()

# Decompose
decomposition = seasonal_decompose(series, model='additive', period=12)  # Yearly seasonality

# Plot components
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 8))
decomposition.observed.plot(ax=ax1); ax1.set_title('Observed')
decomposition.trend.plot(ax=ax2); ax2.set_title('Trend')
decomposition.seasonal.plot(ax=ax3); ax3.set_title('Seasonal')
decomposition.resid.plot(ax=ax4); ax4.set_title('Residual')
plt.tight_layout()
plt.show()
```

```python
##Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results. (Include your Python code and output in the code box below.)

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
import statsmodels.datasets as datasets

# Load data
data = datasets.get_rdataset("AirPassengers", "datasets")
df = pd.DataFrame(data.data)
df['time'] = pd.date_range(start='1949-01-01', periods=len(df), freq='M')
df.set_index('time', inplace=True)
series = df['value']

# Train SARIMA (example order: (1,1,1)(1,1,1,12))
model = SARIMAX(series, order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit()

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_df = forecast.summary_frame()

# Visualize
plt.figure(figsize=(10, 6))
plt.plot(series, label='Original')
plt.plot(forecast_df['mean'], label='Forecast', color='red')
plt.fill_between(forecast_df.index, forecast_df['mean_ci_lower'], forecast_df['mean_ci_upper'], color='pink', alpha=0.3)
plt.title('SARIMA Forecast for AirPassengers')
plt.legend()
plt.show()
```

```python
##Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib. (Include your Python code and output in the code box below.)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import LocalOutlierFactor
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Iris dataset and add synthetic outliers
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add 5 synthetic outliers (extreme values to simulate anomalies)
outliers = pd.DataFrame({
    'sepal length (cm)': [10, 1, 12, 0.5, 15],
    'sepal width (cm)': [5, 1, 6, 0.1, 7],
    'petal length (cm)': [8, 0.5, 10, 0.2, 12],
    'petal width (cm)': [3, 0.1, 4, 0.05, 5]
})
data = pd.concat([data, outliers], ignore_index=True)

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 2: Apply Local Outlier Factor (LOF)
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
data['is_anomaly'] = lof.fit_predict(data_scaled)
data['lof_score'] = -lof.negative_outlier_factor_  # Higher score means more anomalous

# Step 3: Dimensionality reduction for 2D visualization using PCA
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)
data['pca1'] = data_pca[:, 0]
data['pca2'] = data_pca[:, 1]

# Step 4: Visualize anomalies
plt.figure(figsize=(10, 6))
plt.scatter(data[data['is_anomaly'] == 1]['pca1'], data[data['is_anomaly'] == 1]['pca2'],
            color='blue', label='Normal Points')
plt.scatter(data[data['is_anomaly'] == -1]['pca1'], data[data['is_anomaly'] == -1]['pca2'],
            color='red', label='Anomalies', marker='x', s=100)
plt.title('Anomalies Detected by LOF (Projected to 2D PCA Space)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid(True)
plt.show()

# Print summary
print("Number of detected anomalies:", (data['is_anomaly'] == -1).sum())
print("Sample LOF scores for anomalies:\n", data[data['is_anomaly'] == -1]['lof_score'])
```

```python
("""Python Code and Output
Below is example Python code demonstrating anomaly detection with Isolation Forest and short-term forecasting with SARIMAX on synthetic data
(mimicking 15-minute intervals). In a real setup, replace with actual streaming data ingestion.""")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

# Step 1: Generate synthetic data (15-min intervals for 2 days)
np.random.seed(42)
dates = pd.date_range(start='2025-07-01', periods=192, freq='15T')  # 192 points = 2 days
data = pd.DataFrame({
    'timestamp': dates,
    'region': np.random.choice(['North', 'South'], 192),
    'temperature': np.random.normal(25, 5, 192),  # Weather condition
    'energy_usage': 100 + np.sin(np.arange(192) * np.pi / 48) * 20 + np.random.normal(0, 5, 192)
})
# Add anomalies: spikes and drops
data.loc[50:55, 'energy_usage'] += 50  # Spike
data.loc[100:105, 'energy_usage'] -= 40  # Drop

# Preprocess: Set index, encode region
data.set_index('timestamp', inplace=True)
data['region_encoded'] = data['region'].map({'North': 0, 'South': 1})

# Step 2: Anomaly Detection with Isolation Forest
features = ['energy_usage', 'temperature', 'region_encoded']
iso_forest = IsolationForest(contamination=0.05, random_state=42)
data['anomaly_score'] = iso_forest.fit_predict(data[features])
data['is_anomaly'] = data['anomaly_score'] == -1

# Visualize anomalies
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['energy_usage'], label='Energy Usage')
plt.scatter(data[data['is_anomaly']].index, data[data['is_anomaly']]['energy_usage'], color='red', label='Anomalies')
plt.title('Energy Usage with Detected Anomalies (Isolation Forest)')
plt.xlabel('Timestamp')
plt.ylabel('Energy Usage')
plt.legend()
plt.show()

# Step 3: Short-term Forecasting with SARIMAX
# Resample to 15-min (already is), use energy_usage as endogenous, temperature as exogenous
train = data.iloc[:150]
test = data.iloc[150:]
exog_train = train['temperature']
exog_test = test['temperature']

model = SARIMAX(train['energy_usage'], exog=exog_train, order=(1,1,1), seasonal_order=(1,1,1,96))
model_fit = model.fit(disp=False)
forecast = model_fit.forecast(steps=len(test), exog=exog_test)

# Validate
mae = mean_absolute_error(test['energy_usage'], forecast)
print(f"Forecast MAE: {mae:.2f}")

# Visualize forecast
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['energy_usage'], label='Train')
plt.plot(test.index, test['energy_usage'], label='Test')
plt.plot(test.index, forecast, label='Forecast', color='green')
plt.title('SARIMAX Forecast of Energy Usage')
plt.xlabel('Timestamp')
plt.ylabel('Energy Usage')
plt.legend()
plt.show()

# Simulated Monitoring: Check for drift (simple example)
drift_stat = np.abs(train['energy_usage'].mean() - test['energy_usage'].mean())
print(f"Mean Drift: {drift_stat:.2f} (Alert if >5)")
```
